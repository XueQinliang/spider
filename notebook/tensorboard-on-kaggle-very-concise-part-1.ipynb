# Tensorboard Tutorial - Very Concise - Part 1

In this part, you will learn the basics of tensorbaord. Topics covered are as follows:
* Visualizing very simple tf computation graph on tensorboard
* Adding names to the components of computation graph
* Creating scopes to bundle some part of computation graph
* How to visualizing using tf-board on Kaggle

Notaions :-
* tf - Tensorflow

```python
import tensorflow as tf # This is how we import tf
```

## 1. Start with a very basic tf code

```python
# Step 1 : we have created computation graph
# Here we have defined 2 constants named , x and y
x = tf.constant(10)
y = tf.constant(20)

# f = x + y
f = tf.add(x,y)

# Step 2 : we have created tf session and evaluated the value of f
with tf.Session() as sess:
    print(sess.run(f))
    # Now we are going to visualize this simple computation graph
    # Here we are saving computation graph in folder named "folder_to_save_graph_1"
    writer = tf.summary.FileWriter("folder_to_save_graph_1", sess.graph)
    writer.close()
```

### 1.1 Opening tf-board on your local machine 
Now we will run `tensorboard --logdir=path_to_log_directory` to start tf-board. In this particular case, we will run `tensorboard --logdir=folder_to_save_graph_1`
For opening on kaggle refer to section 4

## 2. Adding names

```python
# Step 1 : we have created computation graph
# Here we have defined 2 constants named , x and y
x = tf.constant(10, name="x_constant")
y = tf.constant(20, name='y_constant')

# f = x + y
f = tf.add(x,y,name='function_f')
# Step 2 : we have created tf session and evaluated the value of f
with tf.Session() as sess:
    print(sess.run(f))
    # Now we are going to visualize this simple computation graph
    # Here we are saving computation graph in folder named "folder_to_save_graph_1"
    writer = tf.summary.FileWriter("folder_to_save_graph_2", sess.graph)
    writer.close()
# run command tensorboard --logdir=folder_to_save_graph_2 to see the graph generated by this code
```

## 3. Creating scopes

```python
#Here we are creating scopes for easier visualization
with tf.name_scope("Global"):
    with tf.name_scope("Local_X"):
        # Here we have defined 2 more constants and a function
        x0 = tf.constant(10, name="x0_constant")
        x1 = tf.constant(20, name='x1_constant')
        # fx = x0 + x1
        fx = tf.add(x0,x1,name="x0_plus_x1")
    with tf.name_scope("Local_Y"):
        # Here we have defined 2 more constants and a function
        y0 = tf.constant(50, name="y0_constant")
        y1 = tf.constant(60, name="y1_constant")
        # fy = y0 + y1
        fy = tf.add(y0,y1,name="y0_plus_y1")
    # fg = fx * fy
    fg = tf.multiply(fx,fy,name="fx_into_fy")

# Here we have defined 2 more constants and a function
z0 = tf.constant(100,name="z0_constant")
z1 = tf.constant(100,name="z1_constant")
# fz = z0 + z1
fz = tf.add(z0,z1,name="z0_plus_z1")
# result = fg + fz
result = tf.add(fg,fz,name="result")
    
# Here we are visualizing our scope added graph
with tf.Session() as sess:
	writer = tf.summary.FileWriter("folder_to_save_graph_3", sess.graph)
	print(sess.run(fg))
	writer.close()
# run command tensorboard --logdir=folder_to_save_graph_3 to see the graph generated by this code
```
