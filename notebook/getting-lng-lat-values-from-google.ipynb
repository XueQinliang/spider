## Getting Lng, Lat Values from Google ##

The following shows how to get lng, lat values from Google's API.  It does require a Google API key. Instructions for getting the key can be found [here][1].

Quick sample code is shown below for a few values.  Note, that this won't run in the kernel because calls need to go out to Google.  Plus, a valid key is needed.

I've tried to update all the values, which can be found at the following links:

https://storage.googleapis.com/montco-stats/Larion/2015_lat_lng.csv

https://storage.googleapis.com/montco-stats/Larion/2016_lat_lng.csv


    import urllib.request
    import urllib.parse
    import json
    
    # Need to get Google key 
    #  Reference: https://developers.google.com/maps/documentation/geocoding/get-api-key
    KEY='AIza...'
    
    import pandas as pd
    
    
    # Returns lng,lat
    def getLL(street,city,state):
        address = "%s, %s, %s" % (street,city,state)
        address=urllib.parse.quote_plus(address)
        url = 'https://maps.googleapis.com/maps/api/geocode/json?address=%s=&key=%s' % (address,KEY)
        response = urllib.request.urlopen(url)
        str_response = response.read().decode('utf-8')
        obj = json.loads(str_response)
        m=obj['results'][0]['geometry']['location']
        return (m['lng'],m['lat'])
