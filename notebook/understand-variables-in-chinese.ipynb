# bureau.csv
1. 数据量及描述
	- (1716428,17) 
	- 描述：所有客户之前由其他金融机构提供给信用局的信用报告（对于我们样本中有贷款的客户）。对于我们样本中的每笔贷款，客户在申请日期之前在信用局拥有的信贷数量与行数一样多。
2. SK_ID_CURR
	- 含义：样本中贷款ID (样本中贷款可以有一至多笔在信用局相关的信贷记录)
	- 字段：100002/100003...
3. SK_ID_BUREAU
	- 含义：与我们贷款相关的以前信用局信贷ID
	- 字段：5714462/5714463...
4. CREDIT_ACTIVE
	- 含义：信用局记录在案的信贷的当前状态
	- 字段：Closed/Active/Sold/Bad Debt
5. CREDIT_CURRENCY
	- 含义：信用局信贷的货币单位
	- 字段：currency 1/currency 2/currency 3/currency 4
6. DAYS_CREDIT
	- 含义：客户申请信用局信贷距离当前申请天数
	- 字段：-2922至0
	- 处理方法: **取绝对值**
7. CREDIT_DAY_OVERDUE
	- 含义：根据客户申请贷款的时间点来说，信用局信贷违约天数
	- 字段：0至2792
8. DAYS_CREDIT_ENDDATE
	- 含义：根据客户申请贷款的时间点来说，在信用局信贷剩余的期限
	- 字段：-42,060至31,199 (**正数代表信贷未到期，负数代表信贷已经到期**)
	- 处理方法: **小于-20000值替换为np.nan (不可能在115年前该客户有贷款)**
9. DAYS_ENDDATE_FACT
	- 含义：根据客户申请贷款的时间点来说，信用局信贷已经结束天数(仅限已结束的信贷)
	- 字段：-42,023至0 (42023=115年，肯定是错误的)
	- 处理方法: **小于-4000值替换为np.nan**
10. AMT_CREDIT_MAX_OVERDUE
