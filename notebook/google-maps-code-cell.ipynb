# Google Map Inside Code Cell

The following code show how to display a Google map inside of a code cell. You won't see the map until it's published.  Creating the map itself requires a second Kernel.  Hence, you're creating a Google map in one Kaggle Kernel, then, displaying in an iframe code cell.

You can find the code for the Kaggle Kernel map [here](https://www.kaggle.com/mchirico/d/mchirico/montcoalert/cheltenham-ems-vehicle-accident-map)

To get the height from a given width, multiply by 0.75

700 (width) * 0.75 = 525 (height)


```python
import IPython
url = 'https://www.kaggle.io/svf/430048/d17e3285f78ead25941530f06ee35172/output.html'
iframe = '<iframe src=' + url + ' width=700 height=525></iframe>'
IPython.display.HTML(iframe)
```
